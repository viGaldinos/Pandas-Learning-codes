```python
def optimize_data(data):
    df_data = pd.DataFrame.from_dict(data, orient='index')
    df_data = df_data.transpose()
    df_data = df_data.dropna()
    lista = []
    for i in df_data.keys():
        for j in range(len(df_data.values)):
            lista.append(df_data[i][j])
    df_data1 = pd.DataFrame(lista)

    # Initialize empty dictionary to store the final data
    dic = {}
    for i in df_data.keys():
        # Initialize empty list to store the data for each uuid
        lista4 = []
        for j in df_data1.values:
            # Only append the value if it's timestamp is present in the original data
            if j[0] in [x['t'] for x in data[i]]:
                lista4.append(f"'t': '{j[0]}', 'v': '{j[1]}'")
        dic[i] = lista4
    return dic
```

```python
def pid(data):
    df = pd.DataFrame(columns=['timestamp'])
    for key in data.keys():
        df_temp = pd.DataFrame.from_dict(data[key])
        df_temp = df_temp.rename(columns={'v':key, 't':'timestamp'}).set_index(['timestamp'])
        df = pd.merge(df, df_temp, on='timestamp', how='outer')
    # list of all auxiliar match columns
    match_columns = []
    # identidy matchs
    for col in df.columns:
        if col!='timestamp':
            df[f'match_{col}'] = df[col] == df[col].shift()
            match_columns.append(f'match_{col}')
    df = df.drop(
    df[
        (df[match_columns[0]] == True) 
        & (df[match_columns[1]] == True)
        # add 3rd line if there is one more uuid to compare
        & (df[match_columns[2]] == True)
        ].index)
    # remove auxiliar columns
    df = df.drop(columns=[col for col in match_columns])
    return df

```
