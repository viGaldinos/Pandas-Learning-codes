```python
dias = pd.date_range('20200101', periods = 12)
dias
```

```python
pessoa = ['George', 'Victor', 'Lucas']
```

```python
np.random.choice(pessoa)
```

```python
nome = []
gasto = []
for i in range(12):
  nome.append(np.random.choice(pessoa))
  gasto.append(np.round(np.random.rand() * 100, 2))
nome
```

```python
gasto
```

```python
df = pd.DataFrame({'Dia': dias, 'Nome': nome, 'Gasto': gasto})
df
```

```python
df.pivot(index = 'Dia', columns= 'Nome', values= 'Gasto')
```

```python
carros = [7, 4, 3, 2, 8]
dias = pd.date_range('20190101','20190101', periods= 5)
vendedor = ['George', 'Vagner', 'Pedro', 'Vagner', 'George']

df = pd.DataFrame({
            'Vendas': carros,
            'Data': dias,
            'Vendedor': vendedor
    })
df
```

```python
pd.pivot_table(df, index = 'Data', columns = 'Vendedor', values = 'Vendas', aggfunc= 'sum')
```

```python
data = {
    "localizacao":['CidadeA', 'CidadeB'],
    "Temperatura" : ["Prevista", "Atual"],
    "Set-2019":[30, 32],
    "Out-2019":[45, 43],
    "Nov-2019":[24, 22]
}
print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
pd.melt(df, id_vars=['localizacao'], value_vars=['Temperatura', 'Set-2019'])
```

```python
pd.melt(df, id_vars=['localizacao'], value_vars=['Temperatura', 'Set-2019'], var_name='Var_teste', value_name= 'Nome_do_valor')
```

```python
df = pd.DataFrame(data, columns=['localizacao', 'Temperatura','Set-2019', 'Out-2019', 'Nov-2019'])
print(df)
```

```python
df2 = pd.melt(df, id_vars=["localizacao", "Temperatura"], var_name="Data", value_name="Valor")
print(df2)
```

```python
df2 = pd.melt(df, id_vars=["localizacao"], var_name="Data", value_name="Valor")
print(df2)
```
