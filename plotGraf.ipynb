```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('fast')
```

```python
plt.rcParams['figure.figsize'] = (20, 7)
```

```python
d = pd.read_csv('titanic_train.csv')
d.head()
```

```python
plt.plot(d.Age, "*-r")
plt.xlabel('Passageiro', size = 18)
plt.ylabel('Idade', size = 18)
plt.title('Informações dos passageiros do Titanic')
plt.show()
```

```python
d.Age.plot()
plt.xlabel('Passageiro', size = 18)
plt.ylabel('Idade', size = 18)
plt.show()
```

```python
d.plot()
plt.show()
```

```python
#Scatter plot
plt.scatter(d.PassengerId, d.Age, color = 'yellow', marker= '*')
plt.show()
```

```python
d.Age.hist()
plt.xlabel('Idade', size =18)
plt.ylabel('Frequência observada', size =18)
plt.savefig('nome_fig.png', transparent = True)
plt.show()
```
